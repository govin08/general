```python
def get_hstr(path):
    hstr = pd.read_csv(path)
    hstr['datetime'] = pd.to_datetime(hstr['CLCT_UNIX_TM'], unit='s', utc=True)
    hstr['datetime'] = hstr['datetime'].dt.tz_convert('Asia/Seoul').dt.strftime('%Y-%m-%d %H:%M:%S')
    hstr['datetime'] = hstr['datetime'].astype('datetime64[s]')
    hstr.columns = ['현장교차로ID', '수집유닉스시각', '수집일시', '제어구분코드', '제어상태코드', '주기시간', '옵셋시간',
        'A링1현시시간', 'A링2현시시간', 'A링3현시시간', 'A링4현시시간', 'A링5현시시간', 'A링6현시시간',
        'A링7현시시간', 'A링8현시시간', 'B링1현시시간', 'B링2현시시간', 'B링3현시시간', 'B링4현시시간',
        'B링5현시시간', 'B링6현시시간', 'B링7현시시간', 'B링8현시시간', 'A링1현시보행시간',
        'A링2현시보행시간', 'A링3현시보행시간', 'A링4현시보행시간', 'A링5현시보행시간', 'A링6현시보행시간',
        'A링7현시보행시간', 'A링8현시보행시간', 'B링1현시보행시간', 'B링2현시보행시간', 'B링3현시보행시간',
        'B링4현시보행시간', 'B링5현시보행시간', 'B링6현시보행시간', 'B링7현시보행시간', 'B링8현시보행시간',
        '수집날짜시각']
    cols = list(hstr.columns).copy()
    cols.remove('수집날짜시각')
    cols.insert(1, '수집날짜시각')
    hstr = hstr[cols]
    return hstr
```

```python
from datetime import datetime

inter2trans_times = {5031:['06:00', '06:30', '07:00', '09:30'], # 15
                     5032:['06:00', '06:30', '07:00', '09:30'], # 169
                     5033:['06:00', '06:30', '07:00', '09:30'], # 98
                     5034:['05:30', '06:30', '07:00', '09:30'], # 27
                     5035:['05:30', '06:30', '07:00', '09:30'], # 81
                     5036:['06:00', '06:30', '07:00', '09:30'], # 43
                     5037:['06:00', '06:30', '07:00', '09:30'], # 810
                     5038:['06:00', '06:30', '07:00', '09:30'], # 810
                     5039:['06:00', '06:30', '07:00', '09:30'], # 43
                     5040:['06:00', '06:30', '07:00', '09:30'], # 43
                     5041:['06:00', '06:30', '07:00', '09:30'], # 43
                     5042:['06:00', '06:30', '07:00', '09:30'], # 43
                     5043:['06:00', '06:30', '07:00', '09:30'], # 43
                     5044:['06:00', '06:30', '07:00', '09:30'], # 43
                     5045:['05:00', '05:30', '06:00', '09:30'], # 26
                     5046:['05:00', '05:30', '06:00', '09:30'], # 26
                     5047:['05:00', '05:30', '06:00', '09:30']} # 26

hstrs = {}
diffs = {}
erros = {}
for inter_no in inter2trans_times:
    print(inter_no)
    hstr_temp = hstr[hstr['현장교차로ID']==inter_no].sort_values(by='수집날짜시각').reset_index(drop=True)
    hstr_temp = hstr_temp[['현장교차로ID', '수집날짜시각', '주기시간', '옵셋시간',
                           'A링1현시시간', 'A링2현시시간', 'A링3현시시간', 'A링4현시시간', 'A링5현시시간', 'A링6현시시간']]
    trans_times = inter2trans_times[inter_no]
    hstrs[inter_no] = {}
    diffs[inter_no] = {}
    erros[inter_no] = {}
    for trans_time in trans_times:
        hstrs[inter_no][trans_time] = {}
        diffs[inter_no][trans_time] = {}
        erros[inter_no][trans_time] = {}
        print(trans_time)
        hour, minute = trans_time.split(":")
        hour, minute = int(hour), int(minute)
        df12 = hstr_temp[(hstr_temp['수집날짜시각']>=datetime(2023, 12, 12, hour, minute))&(hstr_temp['수집날짜시각']<datetime(2023, 12, 12, hour, minute+29))]
        df13 = hstr_temp[(hstr_temp['수집날짜시각']>=datetime(2023, 12, 13, hour, minute))&(hstr_temp['수집날짜시각']<datetime(2023, 12, 13, hour, minute+29))]
        hstrs[inter_no][trans_time][12] = df12
        hstrs[inter_no][trans_time][13] = df13
        consist12 = df12.iloc[-1].수집날짜시각
        consist13 = df13.iloc[-1].수집날짜시각
        diff = int((consist12+timedelta(1)-consist13).total_seconds())
        diffs[inter_no][trans_time] = diff
        print(f'안정화된 수집시각의 차이 : {diff}')
        error = int(np.linalg.norm(df12.iloc[:6,4:].values - df13.iloc[:6,4:].values, ord=1))
        erros[inter_no][trans_time] = error
        print(f'처음 여섯 현시시각의 차이(L1 norm) : {error}')
        print(error)
        display(df12)
        display(df13)
```
